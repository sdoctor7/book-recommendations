```python
def create_year_bins(x):
    if np.isnan(x):
        return np.nan
    x=int(x)
    if x<=0:
        return 'BC'
    elif x<= 1760:
        return 'Pre-Industrial'
    elif x<= 1840:
        return 'Industrial Revolution'
    elif x<= 1950:
        return 'Early Modern'
    elif x< 2000:
        return 'Late Modern'
    else:
        return 'Contemporary'

def normalize_year(x):
    if np.isnan(x):
        return np.nan
    if int(x)<1700:
        x=1700
    return 1-((2017-x)/318.0)
```

```python
2017-1700
```

```python
books_meta['Time_Categorization']=books_meta.original_publication_year.apply(create_year_bins)
```

```python
books_meta[['Time_Categorization','original_publication_year']].head()
```

```python
books_meta['Normalized_Year']=books_meta.original_publication_year.apply(normalize_year)
```

```python
def format_language_code(x):
    if pd.isnull(x):
        return np.nan
    elif x in ['eng','en-US','en-GB','en-CA']:
        return 'eng'
    elif x not in ['ara','fre','ind','spa','ger']:
        return 'other'
    else:
        return x
```

```python
books_meta['language_category']=books_meta.language_code.apply(format_language_code)
```
